# Ejercicios



El dataset `BlackFriday.csv` contiene 550.000 registros sobre las ventas realizadas en un comercio. Cada registro contiene información del cliente (sexo, si está casado o no, edad, etc.) e información de la compra que ha realizado en el comercio (importe de la compra, tipo de producto, etc.).

* `User_ID`                : Código de usuario
* `Product_ID`             : Código de producto
* `Sexo`                   : Hombre(M), Mujer(F)
* `Edad`                   : Franja de edad del usuario
* `Occupation`              
* `Cat_ciudad`             : Categoría de la ciudad de residencia del usuario
* `Years_en _la_ciudad`    : Número de años que el usario lleva residiendo en la ciudad
* `Estado_civil`           : Casado(1), soltero(0)
* `Product_Category`
* `Imp_Compra`             : Importe o precio del producto



### 1. 
Carga los datos del  dataset  `BlackFriday.csv`  en un dataframe llamado `ventas` .

```python
import pandas as pd
ventas = pd.read_csv('./BlackFriday.csv')
ventas
```

### 2. 
Calcula los primeros estadísticos `mínimo`, `máximo`, `media` y `desviación típica` del importe de compra  `Imp_Compra`.


```python

ventas['Imp_Compra'].min()
ventas['Imp_Compra'].max()
ventas['Imp_Compra'].mean()
ventas['Imp_Compra'].std()
```

###  3.
¿Cuántas unidades se ha vendido por rango de edad? Usa la operación `value_counts()`.


```python
ventas['Edad'].value_counts()
```

### 4
¿Cuantos valores `NaN` hay en la columna `Product_Category`? Usa la operación `isnull`.

```python
pd.isna(ventas['Product_Category']).sum()
```

### 5.
Sustituye los valores `NaN` de la columna `Product_Category` por el valor `'Desconocido'`.

```python
ventas['Product_Category'].fillna('Desconocido', inplace=True)
ventas['Product_Category']
```

### 6
Selecciona las filas de las compras realizadas por hombres solteros. Guarda el resultado de la selección anterior  en otro DataFrame llamado `compras_solteros`.


```python
compras_solteros = ventas[(ventas['Sexo'] == 'M') & (ventas['Estado_civil'] == 0)]

compras_solteros
```

###  7 
Selecciona a los usuarios que han realizado alguna compra con importe por encima de la media. 


```python
compras_media = ventas[ventas['Imp_Compra'] > ventas['Imp_Compra'].mean()]

compras_media
```

### 8
Añadir una nueva columna al dataframe llamada `Imp_final` que contenga el importe total de la compra al que se le ha aplicado un descuento del 10% (Importe de la compra menos el 10%).

```python
ventas['Imp_final'] = ventas['Imp_Compra'] * 0.9

ventas
```

### 9
Calcula el número de compras realizadas por mujeres y el número de compras realizadas por hombres.

```python
compras_por_genero = ventas['Sexo'].value_counts()

compras_por_genero
```

###  10
Calcula la media del importe. Crear una nueva columna en el dataframe `ventas` llamada 'Tipo_venta' cuyo valor sea 'Low' si el importe de la compra está por debajo de la media y 'High' si el valor de la compra es mayor que la media. Usa la operación `map` sobre la columna ìmporte para aplicar una función `lambda` que haga la transformación.


```python
media_importe = ventas['Imp_Compra'].mean()

ventas['Tipo_venta'] = ventas['Imp_Compra'].map(lambda x: 'LOW' if x < media_importe else 'HIGH')

ventas
```
